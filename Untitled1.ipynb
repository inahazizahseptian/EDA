```python
import pandas as pd
import numpy as np

data = pd.DataFrame(
{
"Negara": ["Indosenia", "Malaysia", "Singapura", "Malaysia", "Thailand", None],
"A": [1, 2, None, 2, 3, None],
"B": [5, None, 7, None, None, None],
"C": [99, 9, None, 9, 11, None],
"D": [12, 8, 14, 8, 15, None]
}
)
#Print dataset
print("Original dataset:")
print(data)
```

```python
data.describe()
```

# **Missing Values**
Lihat pada data di atas, terlihat bahwa ada nilai yang NaN atau kosong. Dapat diperiksa jumlah nilai yang hilang atau kosong pada tiap kolom data dengan fungsi isnull() pada library Pandas di Python:

```python
print("\nMissing values in the entire dataset:")
print(data.isnull().sum())
```

```python
print("\nColumns with missing values:")
print(data.columns[data.isnull().any()].tolist())
```

# Handling Missing **Values**


*   Dihapus

Untuk menangani data null kita dapat langsung menghapus baris yang semuanya
adalah data null dengan fungsi dropna() pada library Pandas:



```python
drop_row = data.dropna(how='all')
drop_row
```

```python
drop_row = data.dropna(how='any')
drop_row
```

```python
drop_col = data.dropna(thresh = 4, axis='columns')
drop_col
```



*   Diisi dengan nilai tertentu

Kita dapat mengisi nilai kosong atau null dengan fungsi fillna() pada library Pandas. Misal, kita ingin mengisi seluruh data kosong dengan nilai rerata atau mean, maka:


```python
fill_na = data.fillna(
    {
        "A" : data["A"].mean(),
        "B" : data["B"].mean(),
        "C" : data["C"].mean(),
        "D" : data["D"].mean()
        }
    )
fill_na
```

```python
#Forward missing values in coloumn B
data_imputed = data["B"].fillna(method="ffill")
print("\nForward-filled column B")
print(data_imputed)


#Forward missing values in coloumn B
data_imputed = data["B"].fillna(method="bfill")
print("\nBackward-filled column B")
print(data_imputed)
```

```python
print("After interpolation")
print(data.interpolate())
```

# **Typo**

Bila diperhatikan kembali, pada kolom “Negara” baris pertama terdapat typo ejaan, yaitu “Indosenia”. Kita dapat mengganti typo tersebut dengan fungsi replace():

```python
typo_data = data.copy ()
typo_data["Negara"] = typo_data["Negara"].replace(["Indosenia"], "Indonesia")
typo_data
```

# Outlier

Untuk melihat visualisasi data outlier, kita gunakan fungsi boxplot():

```python
boxplot = data.boxplot(column=['A', 'B', 'C', 'D'])
boxplot
```

```python
drop_row = data.dropna(how='all')
drop_row
```

```python
fill_na = drop_row.fillna({'A':0,'B':0,'C':0,'D':0})
fill_na
```

```python
Q1 = fill_na['C'].quantile(0.25)
Q3 = fill_na['C'].quantile(0.75)
IQR = Q3-Q1
```

```python
upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR
upper_limit
```

```python
new_df = fill_na[fill_na['C'] < upper_limit]
new_df
```

```python
boxplot = new_df.boxplot(column=['C'])
boxplot
```

```python
new_df_cap = fill_na.copy()
new_df_cap['C'] = np.where(
    new_df_cap['C'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['C'] < lower_limit,
        lower_limit,
        new_df_cap['C']
    )
)
new_df_cap
```
